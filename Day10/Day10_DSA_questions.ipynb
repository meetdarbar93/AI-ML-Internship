Move all zeros to the end of the array without changing the relative order of non-zero elements.

Example: [0,1,0,3,12] → [1,3,12,0,0]



```python
lst = [0,3,4,5,0,0,23,5,0]
for i in lst:
  if i == 0:
    lst.remove(i)
    lst.append(i)
print(lst)
```

Check if an array is a rotation of another array.

 Example: [1,2,3,4,5] rotated → [4,5,1,2,3]

```python
def chk_answer(arr1, arr2):
  if len(arr1) != len(arr2):
    return False
  else:
    str1 = str2 = ""
    for i in range(len(arr1)):
      str1 += str(arr1[i])
      str2 += str(arr2[i])

    str2 = str2 + str2
    return str1 in str2

lst1 = [1,2,3,4,5]
lst2 = [4,5,1,2,3]

ans = chk_answer(lst1,lst2)
if ans:
  print("list 2 is rotation of list 1")

else:
  print("list 2 is not rotation of list 1")
```

Find the first non-repeating element in an array.

 Example: [4,5,1,2,0,4,5,1] → 2

Compute the frequency of each element using a dictionary (unordered map).

```python
def chk_for_repeating(arr):
  freq = dict()
  for i in arr:
    freq[i] = freq.get(i,0)+1

  for num in arr:
    if freq[num] == 1:
      return num

  return None
lst = [4,5,1,2,0,4,5,1]
ans = chk_for_repeating(lst)
print("First non repeated element is",ans)
```

Check if a string has all unique characters (no duplicates).

Example: "abcde" → true, "hello" → false

```python
def chk_string(txt):
  return len(txt) == len(set(txt))
print(chk_string("abcdef"))
print(chk_string("ababab"))
```

Return the longest word in a sentence.

```python
def longest(string):
  arr = string.split()
  longest_word = ""

  for word in arr:
    if len(word) > len(longest_word):
      longest_word = word
  return longest_word

string = "this is example sentence for above problem "
print("Longest word is :",longest(string))
```

Reverse words in a sentence without reversing characters.

"I love coding" → "coding love I"

```python
def rev_word(txt):
  ans = ""
  arr = txt.split()
  for i in range(len(arr)-1,-1,-1):
    ans = ans + arr[i] + " "
  return ans

txt = "i love food"
print(rev_word(txt))
```

Check if a string can be rearranged to form a palindrome.

 Example: "racecar" → yes, "abc" → no

```python
def chk_str(string):
  arr = {char : string.count(char) for char in string}

  odd = 0
  for i in arr.values():
    if i%2 != 0:
      odd += 1

  return True if odd<=1 else False

string = "abcabcd"
if chk_str(string):
  print("String can be possiable for palindrome")
else:
  print("String cannot be possiable for palindrome")
```

Find the longest common prefix in a list of strings.

 Example: ["flower","flow","flight"] → "fl"

```python
def chk_arr(arr):
  prefix = arr[0]
  for i in arr[1:]:
    while not i.startswith(prefix):
      prefix = prefix[:-1]
      if prefix == "":
        return None
  return prefix

lst = ["meet","meeting","meter","mentos"]

print("longest common prefix in list :",chk_arr(lst))
```
