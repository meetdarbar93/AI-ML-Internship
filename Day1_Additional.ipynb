#Reverse a string without using slicing



```python
def reverse(text):
  rev_str = ''
  for i in text:
    rev_str = i + rev_str
  return rev_str

text = input('enter string : ')
rev_str = reverse(text)
print(f'Original String {text}')
print(f'Reversed String {rev_str}')
```

#Count frequency of elements in a list


```python
def freq(lst):
  dict1 = {}
  for i in lst:
    if i in dict1:
      dict1[i] += 1
    else:
      dict1[i] = 1
  return dict1

lst = [1,2,2,3,4,5,3,4,2,1,4,5]
ans = freq(lst)
print(ans)
```

# Find the largest and smallest number in a list
Without using the built-in max() and min() functions.


```python
def find_min_max(lst):
  min = lst[0]
  max = lst[0]
  for i in lst:
    if i < min:
      min = i
    if i > max:
      max = i
  return min , max

lst = [10,53,23,77,44,39,90,5,33]
min , max = find_min_max(lst)
print(f'Smallest Number : {min}')
print(f'Largest Number : {max}')
```

#Find factorial of a number using a loop and recursion


```python
#using loop
def factorial(n):
  fact = 1
  for i in range(1,n+1):
    fact = fact * i
  return fact

n = int(input('enter number : '))
fact = factorial(n)
print(f'Factorial of {n} is {fact}')
```

```python
#using recursion
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

n = int(input('enter number : '))
fact = factorial(n)
print(f'Factorial of {n} is {fact}')
```

#Write a function that checks whether the given number is a prime number.


```python
def is_prime(num):
  if num == 0 or num == 1:
    return False
  else:
    for i in range(2,int(pow(num,1/2))+1):
      if num%i == 0:
        return False
    else:
      return True

n = int(input('enter number : '))
if is_prime(n):
  print(f'{n} is a prime number')
else:
  print(f'{n} is not prime number')
```

# bubble sort or selection sort


```python
# bubble sort
nums = [5, 2, 8, 1, 3]

n = len(nums)
for i in range(n):
  for j in range(0,n-i-1):
    if nums[j] > nums[j+1]:
      nums[j], nums[j+1] = nums[j+1], nums[j]  # swap

print("Sorted list:", nums)
```

```python
# selection sort
nums = [5, 2, 8, 1, 3]

n = len(nums)
for i in range(n):
    min_index = i
    for j in range(i + 1, n):
        if nums[j] < nums[min_index]:
            min_index = j
    nums[i], nums[min_index] = nums[min_index], nums[i]  # swap

print("Sorted list:", nums)
```

#Count uppercase, lowercase, digits, and special characters in a string


```python
txt = "Hello@123"

upper = lower = digit = special = 0

for ch in txt:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    elif ch.isdigit():
        digit += 1
    else:
        special += 1

print(f"Upper = {upper}, Lower = {lower}, Digits = {digit}, Special = {special}")
```
