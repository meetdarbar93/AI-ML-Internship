#Final Summary (6–10 Bullet Points)

**K-Means** worked best because it produced clearer cluster boundaries and a higher silhouette score compared to Hierarchical Clustering.

**The Elbow Method** indicated 4 clusters, confirming natural customer segmentation patterns based on income and spending score.

Clusters revealed noticeable patterns such as **high-income high-spenders**, low-income low-spenders, and **moderate customers**, showing meaningful grouping behavior.

**Hierarchical Clustering** captured relationships but showed more overlap; dendrogram analysis suggested less distinct separation than K-Means.

**Silhouette score** validated that **K-Means clusters were more compact and better separated compared to hierarchical clusters**.

**PCA** reduced the dataset to 2 components, keeping the most important variance and improving visualization without affecting clustering accuracy.

**PCA plots** showed clean visual separation for K-Means clusters, supporting that the dataset’s structure fits well into 4 groups.

Unsupervised learning here focused on finding hidden patterns, not accuracy, and interpretation of cluster behavior mattered more than metrics.

PCA helped remove noise and simplify dimensions, making the clusters visually interpretable in 2D.

Overall, **K-Means is the preferred model** due to its strong performance, clearer cluster shapes, and **better alignment with PCA visualization**.
