#Basic plot functions

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10, 20, 15, 30]
plt.plot(x, y,color="red",linestyle="-",linewidth=2,marker= "o",label="2025 data")
plt.title("Basic functions of plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(color="gray",linestyle=":")
plt.xlim(0,5)
plt.ylim(0,50)
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```

```python
day = ['mon','tus','wed','thr','fri']
sell = [10,15,7,20,12]
plt.plot(day,sell)
plt.title('Bakery cake selling report')
plt.xlabel('Day of week')
plt.ylabel('Sells of week')
plt.show()
```

#fill_between graph

```python
#using simple values
import pandas as pd
import matplotlib.pyplot as plt

# Simple DataFrame
df = pd.DataFrame({
    "x": [0, 1, 2, 3, 4, 5],
    "high": [3, 4, 5, 6, 7, 8],
    "low": [1, 2, 3, 4, 5, 6]
})

plt.fill_between(df["x"], df["low"], df["high"], color="lightgreen", alpha=0.5)

plt.title("fill_between using two DataFrame columns")
plt.xlabel("X-axis")
plt.ylabel("Y values")
plt.show()
```

```python
#Fill under a curve
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y = np.sin(x)

plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title("Area Under the Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
#fill between two lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5

plt.plot(x, y1, label='y1 = sin(x)')
plt.plot(x, y2, label='y2 = sin(x) + 0.5')

plt.fill_between(x, y1, y2, color='orange', alpha=0.3)

plt.title("Fill Between Two Curves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#conditional curve
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='black')
plt.fill_between(x, y, where=(y > 0), color='green', alpha=0.4)
plt.fill_between(x, y, where=(y < 0), color='red', alpha=0.4)

plt.title("Conditional Fill: Green for Positive, Red for Negative")
plt.show()

```

#plot.area()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2010, 2021)
df = pd.DataFrame({
    "Petrol":   [75, 72, 70, 67, 63, 60, 55, 50, 46, 43, 40],
    "Diesel":   [50, 52, 54, 55, 53, 50, 47, 45, 43, 40, 37],
    "Electric": [5,  7, 10, 14, 18, 23, 30, 38, 47, 58, 70]
}, index=years)

df.plot.area(stacked=True, alpha=0.7, figsize=(8,5))
plt.title("Car Sales by Fuel Type (20102020)")
plt.xlabel("Year")
plt.ylabel("Number of Cars Sold (in thousands)")
plt.show()
```

#Stackplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = {
    "year": np.arange(2010, 2021),
    "mpg": [25, 26, 27, 29, 30, 31, 33, 34, 36, 37, 39],
    "weight": [1500, 1480, 1450, 1420, 1400, 1380, 1360, 1340, 1320, 1300, 1280]
}

df = pd.DataFrame(data)

plt.stackplot(df["year"], df["mpg"], df["weight"], labels=["MPG", "Weight"], alpha=0.7)

#do same thing using fill_between
# plt.fill_between(df["year"], df["mpg"], color="green", alpha=0.5, label="MPG")
# plt.fill_between(df["year"], df["weight"], color="orange", alpha=0.5, label="Weight")

plt.legend()
plt.title("Car MPG and Weight Over Time (20102020)")
plt.xlabel("Year")
plt.ylabel("Values")
plt.show()
```

#Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Math": [88, 92, 80, 89, 100],
    "Science": [78, 85, 82, 90, 94],
    "English": [90, 88, 85, 87, 93],
    "History": [70, 75, 78, 85, 90],
    "Art": [95, 98, 92, 96, 99]
}

x = pd.DataFrame(data, index=["A", "B", "C", "D", "E"])
sns.heatmap(x, cmap="coolwarm", annot=True, linewidths=0.5,linecolor="black")

plt.title("Students' Subject Scores Heatmap")
plt.show()
```

```python
#with random numbers
tempreature=np.random.randint(low=1,high=100,size=(10,10))
print(tempreature)
sns.heatmap(data=tempreature,annot=True,cmap="coolwarm")
```
