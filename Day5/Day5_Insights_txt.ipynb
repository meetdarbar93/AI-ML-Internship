#Machine Learning

Definition:

Machine Learning is a branch of Artificial Intelligence (AI) that enables computers to learn automatically from data and improve their performance on a task without being explicitly programmed.

In simple terms,

Machine Learning is about feeding data to computers so they can find patterns and make decisions or predictions on their own.

#Types of Machine Learning

Machine Learning can be divided into three main types:

| **Type**          | **Learning Data** | **Goal**                       | **Examples**                             |
| ----------------- | ----------------- | ------------------------------ | ---------------------------------------- |
| **Supervised**    | Labeled Data      | Predict output                 | Spam detection, price prediction         |
| **Unsupervised**  | Unlabeled Data    | Find structure/patterns        | Customer segmentation, anomaly detection |
| **Reinforcement** | Trial & error     | Learn best actions via rewards | Game bots, self-driving cars             |


##1.  Supervised Learning

Definition:

The model is trained on labeled data (data with input-output pairs).
It learns the mapping from input → output.

Goal:
Predict the output for new unseen inputs.

 a) Regression

Predict continuous numeric values.

Example: Predicting house prices, temperature, or stock prices.

Algorithms: Linear Regression, Decision Tree Regression

 b) Classification

Predict categorical/discrete values (classes or labels).

Example: Spam vs Not Spam, Disease detection (Positive/Negative).

Algorithms: Logistic Regression, K-Nearest Neighbors (KNN), Support Vector Machine (SVM)





#Workflow of supervised learning

| **Step**             | **Purpose**                          | **Function / Tool**  |
| -------------------- | ------------------------------------ | -------------------- |
| 1️.   Train-Test Split | Divide data for training and testing | `train_test_split()` |
| 2️.   Model Fitting    | Learn patterns from training data    | `model.fit()`        |
| 3️.   Predictions      | Predict output for test data         | `model.predict()`    |
| 4️.   Evaluation       | Measure accuracy/performance         | `metrics` module     |



#Scikit-learn Basics

| **Concept**  | **Function / Module** | **Purpose**                          |
| ------------ | --------------------- | ------------------------------------ |
| Load dataset | `sklearn.datasets`    | Load built-in or external data       |
| Split data   | `train_test_split()`  | Divide data into train and test sets |
| Train model  | `.fit()`              | Learn from training data             |
| Predict      | `.predict()`          | Predict output for new data          |
| Evaluate     | `sklearn.metrics`     | Measure model performance            |

#terms used in classification_report

| Term          | Meaning                              | Formula         | Goal                    |
| ------------- | ------------------------------------ | --------------- | ----------------------- |
| **Precision** | Correctness of positive predictions  | TP / (TP + FP)  | High = few false alarms |
| **Recall**    | How many actual positives were found | TP / (TP + FN)  | High = few misses       |
| **F1-score**  | Balance between precision & recall   | 2 × (P×R)/(P+R) | High = good balance     |
| **Support**   | Total actual samples                 | count           | Just info               |
