# Chapter: Data Visualization in Python

## 1. Line Plot
Definition:

A line plot connects data points with lines to show trends or changes over time.

Usecase:
Used to visualize trends, patterns, or continuous data over a period.

example:

plt.plot(x, y, color, linestyle, linewidth, marker, label)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.legend()

plt.grid(True)

plt.show()


## 2. Bar Chart
Definition:

Represents categorical data with rectangular bars. The height of the bar shows the value.

Usecase:
Used to compare quantities of different categories.

example:

plt.bar(categories, values, color, label)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.legend()

plt.show()


## 3. Pie Chart
Definition:

Shows the proportion of categories in a whole dataset as slices of a circle.

Usecase:
Used to represent percentage or part-to-whole relationships.

example:

plt.pie(values, labels=labels, autopct='%.2f%%')

plt.title("Title")

plt.show()


## 4. Histogram
Definition:

Shows frequency distribution of numerical data.

Usecase:
Used to analyze distribution, frequency, or range of continuous data.

example:

plt.hist(data, bins, color, edgecolor, label)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Frequency")

plt.legend()

plt.show()


## 5. Scatter Plot
Definition:

Displays relationship between two numerical variables using points.

Usecase:
Used to study correlation, trends, or patterns between variables.

example:

plt.scatter(x, y, color, marker, label)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.legend()

plt.grid(True)

plt.show()


## 6. Fill Between
Definition:

Fills area under a curve or between two curves.

Usecase:
Used to highlight ranges, compare curves, or visualize area under a curve.

example:

plt.fill_between(x, y1, y2=None, where=None, color, alpha)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.show()


## 7. Area Plot
Definition:

Stacked area chart for visualizing composition of totals over time.

Usecase:
Used to visualize contributions of multiple variables over a period.

example:

df.plot.area(stacked=True, alpha)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.show()


## 8. Stackplot
Definition:

Similar to area plot; shows multiple quantities over time stacked.

Usecase:
Used to compare multiple trends cumulatively over time.

example:

plt.stackplot(x, y1, y2, labels=labels, alpha)

plt.title("Title")

plt.xlabel("X-axis label")

plt.ylabel("Y-axis label")

plt.legend()

plt.show()


## 9. Boxplot
Definition:

Shows distribution of data using five statistics: minimum, Q1, median, Q3, maximum; detects outliers.

Usecase:
Used to visualize spread, median, and detect outliers.

Syntax:

plt.boxplot(data)

plt.title("Title")

plt.ylabel("Y-axis label")

plt.show()


## 10. Heatmap
Definition:

Visualizes matrix-like data with colors representing values; good for correlation or intensity.

Usecase:
Used to visualize correlation, patterns, or intensity of values in a table.

example:

sns.heatmap(data, annot=True, cmap="color_map", linewidths=0.5,

linecolor="black")

plt.title("Title")

plt.show()


## 11. Waffle Chart
Definition:

Represents categorical data as squares in a grid; each square corresponds to a value proportion.

Usecase:
Used to display proportion of categories visually in grid format.

example:

from pywaffle import Waffle

plt.figure(FigureClass=Waffle, rows=10, values=data_dict, colors=color_list,

title={'label': 'Title', 'loc': 'center'})

plt.show()
