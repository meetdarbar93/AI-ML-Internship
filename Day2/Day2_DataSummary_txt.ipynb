| Function         | Description                                                                             |
| ---------------- | --------------------------------------------------------------------------------------- |
| `pd.Series()`    | Creates a one-dimensional labeled array (Series) in pandas.                             |
| `pd.DataFrame()` | Creates a 2D data structure (table) with labeled rows and columns.                      |
| `pd.read_csv()`  | Loads a CSV file into a DataFrame.                                                      |
| `.head()`        | Displays the first n rows of a DataFrame.                                               |
| `.info()`        | Shows summary info including columns, data types, and memory usage.                     |
| `.shape`         | Returns a tuple (rows, columns).                                                        |
| `.isnull()`      | Checks for missing (NaN) values.                                                        |
| `.sum()`         | Returns the sum of numeric values (can also count nulls when combined with `isnull()`). |
| `.mean()`        | Calculates the mean of numeric columns.                                                 |
| `.median()`      | Calculates the median value.                                                            |
| `.mode()`        | Returns the most frequent value(s).                                                     |
| `.iloc[]`        | Selects rows/columns by index (integer-based).                                          |
| `.loc[]`         | Selects rows/columns by labels (name-based).                                            |
| `.sort_values()` | Sorts a DataFrame by one or more columns.                                               |
| `.drop()`        | Removes specified rows or columns.                                                      |
| `.describe()`    | Gives summary statistics for numeric columns.                                           |
