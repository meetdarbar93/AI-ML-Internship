#Regular Expression

```python
import re
import pandas as pd

txt = "Hello, Haan Ji Good Morning"
x = r"[A-Z]"
y = re.findall(x,txt)
print(y)
```

```python
txt = "Hello Good MOrning"
x = r"\w{5}" # take 5 character from every words from string
y =   re.findall(x,txt)
print(y)

x = r"\w{5,}" # take 5 or more character from every words from string
y =   re.findall(x,txt)
print(y)

x = r"\b\w{5}\b" # take 5 or more character from every words from string
y = re.findall(x,txt)
print(y)
```

```python
txt = "hello how are you"
x = re.sub(r"\s","$",txt) #Replace all spaces
print(x)

txt = "hello how are you"
x = re.sub(r"\s","$",txt,2) #Replace first two matches
print(x)
```

```python
txt = "hellodbmn 67836 23098"
x = r"\b\d{5}\b" #\d - for digit
y = re.findall(x,txt)
print(y)
```

```python
txt = "hello how are you"
x = re.split(r"\s",txt)
print(x)

txt ="hello how are you"
x = re.split(r"\s",txt,1) # only first find
print(x)
```

```python
#searching
txt="hello how are you"
x=re.search("are",txt)
print(x)
```

```python
txt = "0dadkabbbbjhkjabb780abb0"
x = r"ab+"
y = re.findall(x,txt)
print(y)

x = r"ab*"
y = re.findall(x,txt)
print(y)
x = r"ab?"

y = re.findall(x,txt)
print(y)
```

```python
#extract date
txt = "hello today is bday and date is 12/02/2006"
x = r"\b\d{1,2}/\d{1,2}/\d{4}\b"
y = re.findall(x,txt)
print(y)
```

```python
#to extract email
txt = "hello abc@gmail.com"
x = r"\b[A-Za-z0-9._+%]+@[A-Za-z0-9]+\.[A-Za-z]{2,}\b"

y = re.findall(x,txt)
print(y)
```

```python
#extract url
txt = "http://www.google.com"
x = r"\bhttp://[A-Za-z0-9._]+\.[A-Za-z]{2,}\b"

y = re.findall(x,txt)
print(y)
```

```python
#extract mobile no
txt = "contact me 1234567890 9870645321"
x = r"\b\d{10}\b"
y = re.findall(x,txt)
print(y)
```

```python
#finding patterns
txt="ab abbbb bbbaaabbb ababab bbbbb aaaaa"
x=r"ab*"
y=re.findall(x,txt)
print(y)
```

```python
def one_more_b(txt):
    return bool(re.findall(r"ab+",txt))
print(one_more_b("a"))
print(one_more_b("ab"))
print(one_more_b("abbbb"))
print(one_more_b("bbbaaabbb"))
print(one_more_b("ababab"))
print(one_more_b("bbbbb"))
print(one_more_b("aaaaa"))
```

```python
def one_more_b(txt):
    return bool(re.findall(r"ab+?",txt))
print(one_more_b("ab"))
print(one_more_b("abbbb"))
print(one_more_b("bbbaaabbb"))
print(one_more_b("ababab"))
print(one_more_b("bbbbb"))
print(one_more_b("aaaaa"))
```

```python
def one_more_b(txt):
    return bool(re.findall(r"ab{3}",txt))
print(one_more_b("ab"))
print(one_more_b("abbbb"))
print(one_more_b("bbbaaabbb"))
print(one_more_b("ababab"))
print(one_more_b("bbbbb"))
print(one_more_b("aaaaa"))
```

```python
def one_more_b(txt):
    pat='^[a-z]+_[a-z]+$'
    if re.search(pat,txt):
        return "found match!"
    else:
        return "not found match!"
print(one_more_b("ab"))
print(one_more_b("ab_bbb"))
print(one_more_b("bbb_aaabbb"))
print(one_more_b("ababab"))
print(one_more_b("bb_bbb"))
print(one_more_b("aaaaa"))
```

```python
def one_more_b(txt):
    pat='^[A-Z]+[a-z]+$'
    if re.search(pat,txt):
        return "found match!"
    else:
        return "not found match!"
print(one_more_b("ab"))
print(one_more_b("Abbbb"))
print(one_more_b("bbbaaabbb"))
print(one_more_b("ababab"))
print(one_more_b("bAbbb"))
print(one_more_b("AaaAa"))
```
