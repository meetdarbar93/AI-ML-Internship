#Python re Module – Summary Table
| **Function / Symbol** | **Explanation (One Line)**                                     |
| --------------------- | -------------------------------------------------------------- |
| **re.findall()**      | Returns all non-overlapping matches of the pattern in a list.  |
| **re.search()**       | Searches the string and returns the first matching occurrence. |
| **re.match()**        | Matches a pattern only at the beginning of the string.         |
| **re.fullmatch()**    | Matches only if the entire string fits the pattern.            |
| **re.split()**        | Splits the string using the regex pattern as a delimiter.      |
| **re.sub()**          | Replaces all pattern occurrences with a replacement string.    |
| **re.compile()**      | Pre-compiles a regex pattern for repeated efficient use.       |


#Core regex symbols


| **Symbol** | **Meaning (One Line)**                           |                              |
| ---------- | ------------------------------------------------ | ---------------------------- |
| **.**      | Matches any single character except newline.     |                              |
| **^**      | Matches the start of the string.                 |                              |
| **$**      | Matches the end of the string.                   |                              |
| **[]**     | Matches any one character inside the brackets.   |                              |
| **[^ ]**   | Matches any character *not* inside the brackets. |                              |
| \|| Logical OR between patterns. |
| **()**     | Captures a group of characters or tokens.        |                              |
| **\b**     | Matches a word boundary.                         |                              |
| **\B**     | Matches a non-word boundary.                     |                              |

#Quantifier
| **Quantifier** | **Meaning (One Line)**                     |
| -------------- | ------------------------------------------ |
| **+**          | Matches one or more occurrences.           |
| *          | Matches zero or more occurrences.          |
| **?**          | Matches zero or one occurrence (optional). |
| **{n}**        | Matches exactly *n* occurrences.           |
| **{n,}**       | Matches at least *n* occurrences.          |
| **{n,m}**      | Matches between *n* and *m* occurrences.   |

#Non-Greedy (Lazy) Quantifiers
| **Symbol** | **Meaning (One Line)**                                     |
| ---------- | ---------------------------------------------------------- |
| **+?**     | Matches as few characters as possible, but at least one.   |
| *?     | Matches as few characters as possible, starting from zero. |
| **??**     | Matches zero or one occurrence, preferring zero.           |

#Special character classes
| **Class** | **Meaning (One Line)**                                    |
| --------- | --------------------------------------------------------- |
| **\d**    | Matches any digit (0–9).                                  |
| **\D**    | Matches any non-digit.                                    |
| **\w**    | Matches any word character (letters, digits, underscore). |
| **\W**    | Matches any non-word character.                           |
| **\s**    | Matches whitespace (space, tab, newline).                 |
| **\S**    | Matches non-whitespace characters.                        |
