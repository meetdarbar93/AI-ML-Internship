# Calculating the sum and average of a list of numbers.



```python
def create_list(n):
    lst = []
    for i in range(0,n):
        n = int(input(f'enter element {i+1} : '))
        lst.append(n)
    return lst

def find_sum_avg(lst):
    sum = 0
    for i in lst:
        sum = sum + i
    avg = sum/len(lst)
    print(f"Sum : {sum}")
    print(f"Average : {avg}")

n = int(input('enter total number of list element : '))
lst = create_list(n)
find_sum_avg(lst)
```

# Finding even/odd numbers in a list.


```python
def create_list(n):
    lst = []
    for i in range(0,n):
        n = int(input(f'enter element {i+1} : '))
        lst.append(n)
    return lst

def even_odd_lst(lst):
    even = []
    odd = []
    for i in lst:
        if i%2 == 0:
            even.append(i)
        else:
            odd.append(i)
    print(f"Even Numbers : {even}")
    print(f"Odd Numbers : {odd}")

n = int(input('enter total number of list element : '))
lst = create_list(n)
print(f"Original list {lst}")
even_odd_lst(lst)
```

# Counting vowels in a string.

```python
txt = input('Enter string : ')
vowels = 'aeiouAEIOU'
count = 0
for i in txt:
    if i in vowels:
        count = count + 1
print(f'total vowels {count}')
```

 # Generating Fibonacci series up to n.


```python
n = int(input('enter n : '))
a, b = 0, 1
for i in range(0,n):
  print(a,end=" ")
  a, b = b, a + b
```
