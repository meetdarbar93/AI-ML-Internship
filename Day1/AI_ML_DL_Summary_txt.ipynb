1. Artificial Intelligence (AI)

Definition:
AI (Artificial Intelligence) is a broad field of computer science that aims to create machines or systems capable of performing tasks that normally require human intelligence — such as reasoning, problem-solving, learning, and understanding language.

Key Idea:
AI = Machine that thinks and acts like humans.

Examples:

Chatbots (like ChatGPT or Siri): They understand human language and respond intelligently to user queries.

Self-driving cars: Use AI to make decisions like when to stop, turn, or accelerate.

2. Machine Learning (ML)

Definition:
ML is a subset of AI that focuses on systems that learn from data and improve automatically over time without being explicitly programmed.

Key Idea:
ML = Machine that learns from data.

Examples:

Spam email detection: The system learns from past emails (spam vs. not spam) to classify new emails automatically.

Netflix recommendations: The algorithm learns your viewing habits and suggests movies you’re likely to enjoy.

3. Deep Learning (DL)

Definition:
DL is a subset of ML that uses neural networks with many layers to learn from large amounts of data, especially unstructured data like images, audio, and text.

Key Idea:
DL = Machine that learns complex patterns using neural networks.

Examples:

Face recognition (Facebook tagging, phone unlock): Deep neural networks analyze facial features for identification.

Voice assistants (Google Assistant, Alexa): Use deep learning to understand and generate human speech.

| Feature              | Artificial Intelligence (AI)    | Machine Learning (ML)                   | Deep Learning (DL)                      |
| -------------------- | ------------------------------- | --------------------------------------- | --------------------------------------- |
| **Scope**            | Broad field                     | Subset of AI                            | Subset of ML                            |
| **Goal**             | Mimic human intelligence        | Learn from data                         | Learn from large data using neural nets |
| **Techniques Used**  | Rules, logic, search algorithms | Decision trees, regression, clustering  | Neural networks, CNNs, RNNs             |
| **Data Requirement** | Low to medium                   | Medium                                  | Very high                               |
| **Examples**         | Chatbots, Self-driving cars     | Spam detection, Netflix recommendations | Face recognition, Voice assistants      |
